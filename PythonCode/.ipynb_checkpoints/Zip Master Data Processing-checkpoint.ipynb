```python
import requests
import json
import urllib
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
url = 'https://parseapi.back4app.com/classes/Uszipcode_US_Zip_Code?count=1&limit=43000'
headers = {
    'X-Parse-Application-Id': '', # This is your app's application id
    'X-Parse-REST-API-Key': '' # This is your app's REST API key
}
data = json.loads(requests.get(url, headers=headers).content.decode('utf-8')) # Here you have the data that you need
#print(json.dumps(data, indent=2))

new_data = data['results']
print(len(new_data))

```

```python
us_zip_code = []
types = []
primary_cities = []
states = []
counties = []
timezones = []
area_codes = []
latitudes = []
longitudes = []
countries = []
estimated_pops = []
created_at = []
updated_at = []
for i in new_data:
    us_zip_code.append(str(i['US_Zip_Code']))
    types.append(i['Type'])
    primary_cities.append(i['Primary_city'])
    states.append(i['State'])
    try:
        counties.append(i['County'])
    except:
        counties.append('NA')
    try:
        timezones.append(i['Timezone'])
    except:
        timezones.append('0')
    try:
        area_codes.append(i['Area_codes'])
    except:
        area_codes.append('0')
    latitudes.append(i['Latitude'])
    longitudes.append(i['Longitude'])
    countries.append(i['country'])
    estimated_pops.append(i['estimated_population'])
    created_at.append(i['createdAt'])
    updated_at.append(i['updatedAt'])


```

```python
updated_zip_codes = []
for zip_code in us_zip_code:
    if len(str(zip_code)) == 3:
        updated_zip_codes.append(f'00{zip_code}')
    elif len(str(zip_code)) == 4:
        updated_zip_codes.append(f'0{zip_code}')
    else:
        updated_zip_codes.append(zip_code)
```

```python
zips_df = pd.DataFrame({'us_zip_code': updated_zip_codes,
                          'type': types,
                          'primary_city': primary_cities,
                          'state': states,
                          'latitude': latitudes,
                          'longitude': longitudes,
                          'county': counties,
                          'timezone': timezones,
                          'area_codes': area_codes,
                          'country': countries,
                          'estimated_population': estimated_pops,
                          'created_at': created_at,
                          'updated_at': updated_at,
                       })
zips_df.to_csv('Data/zips_output.csv', index = False, mode = 'w')
zips_df
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/ziproject"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
zips_df.to_sql(name='zips', con=engine, if_exists='append', index=False)
```
