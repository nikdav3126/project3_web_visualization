```python
# Dependencies
import pandas as pd
import os
import csv
```

```python
# Store filepath into a variable
filepath = "Data/Intergenerational_Mobility_csv_2022_04_02.csv"
```

```python
#Read out file with Pandas
df = pd.read_csv(filepath, encoding="UTF-8")
```

```python
# Show header
df.head()
```

```python
df.head()
```

```python
df.reset_index(drop=True)
```

```python
# Remove period in column name, pull spaces out of column names
df.columns = df.columns.str.replace('.',' ')
df.columns = df.columns.str.replace(' ','')
df.head()
```

```python
# Remove extraneous columns
del df['MSAName']
del df['State']
del df['NoofChildreninCoreSample']
del df['RelativeMobilityRank-RankSlope']
df.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
df.to_csv("cleaned_CSVs/CleanedIntergenerationalMobility_2022_04_05.csv", index=False, header=True)
```
